```
# Puesto que el objetivo de este challenge tiene que ver con el "atraso" de los vuelos,
# entonces se hace necesario conocer la diferencia de "timing" entre los vuelos programados
# y la operación de los vuelos.
# De tal forma, creamos una nueva columna dif_seg, que tiene la difencia Fecha-O - Fecha-I en segundos

from datetime import datetime, timedelta 
df['dif_seg']=(pd.to_datetime(df["Fecha-O"],format="%Y-%m-%d %H:%M:%S")-pd.to_datetime(df["Fecha-I"],format="%Y-%m-%d %H:%M:%S"))/timedelta(seconds=1)
df.head(10)
```

Recordar que el objetivo de este desafío dice relación con "predecir la probabildad de atraso de los vuelos", entonces conviene tener una nueva variable que permita saber si el vuelo tiene atraso o no.


```
# Se ve que hay vuelos que adelantan su operación respecto de su programación.
# También se ve que hay vuelos que atrasan su operación respecto de su programación.
# Crearemos una nueva variable "atraso", con valor 1 si hay atraso en el vuelo, y con valor 0, en otro caso.

df['atraso']=df['dif_seg'].apply(lambda x: 1 if x > 0 else 0)
df.head(10)
```

###Ahora abordaremos la obtensión de las tasas de atraso de los vuelos, para las distintas variables: destino, aerolínea, mes del año, día de la semana, temporada, tipo de vuelo.

```
# Se define la nueva función obtiene_tasa(), que dada una variable_pivote y una
# variable_tasa del dataset, retorna un dataset laas tasas de la variable_tasa, 
# respecto de la variable_pivote. En particular, puesto que la variable_tasa, es
# la variable 'atraso', entonces para una variable_pivote, se obtendrán las tasas
# de atraso y de "no atraso"

def obtiene_tasa(df, var_pivote, var_tasa):
  df_aux          = df.groupby([var_pivote,var_tasa]).size().rename('sub_tot').reset_index()
  df_aux['total'] = df_aux.groupby(var_pivote)['sub_tot'].transform('sum')
  df_aux['tasa']  = df_aux['sub_tot']/df_aux['total']
  return df_aux

 
# Para conocer la tasa de atraso de los vuelos respecto de su destino, usaremos
# entonces como variable_pivote a 'Des-O' 
df_dest_tasa = obtiene_tasa(df,'Des-O','atraso')
# Al mirar las primeras 10 muestras del dataset se tiene
print(df_dest_tasa.head(10),'\n') 

# Si se quiere visualizar unicamente la tasa de atraso entonces
print('ESTADISTICOS')
print(df_dest_tasa[df_dest_tasa['atraso']==1][['Des-O','tasa']].rename(columns={'tasa':'tasa_atraso'}).describe(),'\n')
print('TASA DE ATRASO X DESTINO')
df_dest_tasa[df_dest_tasa['atraso']==1][['Des-O','tasa']].rename(columns={'tasa':'tasa_atraso'})
```

Al mirar la tasa de atraso de los vuelos por destino, sus estadísticos muestran que:

1. Tasa de atraso mínimo.        =  0.20

2. Tasa de atraso máximo         =  1.00

3. Tasa de atraso desviación std =  0.18

Sumado a lo anterior, al ver sus percentiles, se podría infererir que la variable Des-O, es relevante en la determinación de la tasa de atraso de un vuelo.

1. El preprocesamiento del dataset.
2. Entrenamiento y prueba de los distintos modelos de clasificación usados.

```
# Como parte del preprocesamiento del dataset, convertiremos alguna vaariables
# categóricas a variables númericas.

from sklearn.preprocessing import OrdinalEncoder

def crea_encode(df,var_cat):

# Creamos el codificador indicandole el orden de la variables
  encoder = OrdinalEncoder(categories=[df[var_cat].unique().tolist()])

# Ajustamos el codificador con la variable OPERA y la transformamos
  encoder.fit(df[[var_cat]])
  df[var_cat + '_encoded'] = encoder.transform(df[[var_cat]])
  return df

df = crea_encode(df,'OPERA')
df = crea_encode(df,'Des-O')
df = crea_encode(df,'periodo_dia')
df
```

```
# Ahora realizaremos la selección de las variables predictoras, y también haremos
# la separación del dataset, en dataset de entrenamiento y dataset de pruebas.
from sklearn.model_selection import train_test_split


# Selección de variables predictoras a ser consideradas
predic_col = ['OPERA_encoded', 'Des-O_encoded', 'MES','DIA']
predic_col = ['OPERA_encoded', 'Des-O_encoded']
predic_col = ['OPERA_encoded']
predic_col = ['OPERA_encoded', 'Des-O_encoded', 'MES','periodo_dia_encoded']
predic_col = ['OPERA_encoded', 'Des-O_encoded', 'periodo_dia_encoded']
predic_col = ['OPERA_encoded', 'Des-O_encoded', 'MES','DIA','periodo_dia_encoded']
target_col = ['atraso']

# Separación de variables predictoras de variable objetivo.
df_X = df[predic_col]
df_Y = df[target_col]

# Split de datos de entrenamiento y de prueba
train_X,test_X,train_Y,test_Y = train_test_split(df_X,df_Y,
                                                 test_size = .25 ,
                                                 random_state = 111)

```

```
# Para este quinto modelamiento, se usa una Red Neuronal, empleando TensorFlow/Keras.
# Se importan las librerias requeridas.
import tensorflow
import keras.utils
from   keras.utils                 import np_utils
from   keras.models                import Sequential
from   keras.layers                import Dense, Dropout, Activation
from   tensorflow.keras.optimizers import  RMSprop

# Se realiza la partición de las muestras para aplicar el entrenamiento de la 
# Red Neuronal.
from sklearn.preprocessing   import LabelEncoder
from sklearn.model_selection import train_test_split


X = df_X.copy()
y = df_Y.values.ravel()
y_cat = keras.utils.np_utils.to_categorical(y, num_classes=2)

X_train_, X_test_, y_train_, y_test_ = train_test_split(X, y_cat, test_size=0.25, stratify=y)
```

###Configuración de la Red Neuronal TensorFlow-Keras RMSprop
1. Una capa de entrada con 64 neuronas, función de activación relu y 5 variables de entrada.
2. Una capa oculta con 32 neuronas,función de activación relu.
3. Una capa de salida con 2 clases y funció de activación softmax.

###Esta configuración es una primera aproximación y se puede ir variando, junto con sus hyperparámetros para lograr mejores resultados. 

```
# Se define el modelo para la Red Neuronal.
# Esta Red Neuronal tiene una capa de entrada con 64 neuronas, una capa oculta
# con 32 neuronas y una capa de salida con 2 neuronas.
model = Sequential()
# La primera capa tiene 64 'neuronas', función de activación relu y 5 variables
# de entrada.
model.add(Dense(64, activation='relu', input_dim=5))
# La segunda capa tiene 32 'neuronas' y función de activación relu.
model.add(Dense(32, activation='relu'))
# La capa de salida, considera las 2 clases y función de activación softmax.
model.add(Dense(2, activation='softmax'))

# Se usa el optimizador RMSprop, con valores de hiperparámetros tomados desde
# la documentación de Keras.
rmsprop = RMSprop(learning_rate=0.001, rho=0.9)
# Se instancia el modelo, usando optimizador rmsprop, función de pérdida 
# categorical_crossentropy y métrica accuracy.
model.compile(optimizer=rmsprop,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
# Se mira el modelo y sus parámetros
model.summary()
# Se entrena la Red Neuronal, con epochs = 100 y batch_size = 64
model.fit(X_train_, y_train_, epochs=100, batch_size=64, verbose=0)
# Se evalúa la Red Neuronal.
score = model.evaluate(X_test_, y_test_, verbose=0)
print('RMSprop')
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
